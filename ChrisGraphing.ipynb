```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
#ax = plt.subplot()

t = np.arange(-2.3, .3, 0.01)
s = 1.6*np.sin(np.power(t,2))
line, = plt.plot(t, s, lw=2)

 
x = np.arange(-.15, .15, 0.01)
x2 = np.arange(-.2, .2, 0.01)
c = .01 #standard deviation
for i in range(len(t)):

    if i % 25 == 0:
        c = .004 #standard deviation
        c = c*random.random()
        a = t[i]
        b = s[i]
        k = x + a
        k2 = x2 + a
        y = 32*np.power(k-a,2)+b
        f = 1.3*np.power(np.e,-(k2-a)*(k2-a)/(2*c)) - 4 #gaussian
        line, = plt.plot(k, y, lw=2, color='orange')
        line, = plt.plot(k2, f, lw=2, color='g')
        


plt.ylim(-5, 2.5)
plt.show()
```
